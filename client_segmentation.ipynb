CLIENT SEGMENTATION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

df = pd.read_csv('walmart.csv')

print(f"Shape del dataset: {df.shape}")

df_clustering = df.drop(['User_ID', 'Product_ID'], axis=1)

categorical_cols = ['Gender', 'Age', 'Occupation', 'City_Category', 'Stay_In_Current_City_Years', 'Marital_Status', 'Product_Category']

encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(df_clustering[categorical_cols])

encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(categorical_cols))

```

```python

numeric_cols = ['Purchase']
df_final = pd.concat([encoded_df, df_clustering[numeric_cols].reset_index(drop=True)], axis=1)

print(f"Shape después de OneHotEncoder: {df_final.shape}")


scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_final)


wcss = []
for i in range(1, 16):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1, 16), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

```

```python

optimal_k = 6
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(df_scaled)

df['Cluster'] = clusters
```

```python


pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_scaled)

plt.figure(figsize=(8,5))
sns.scatterplot(x=principal_components[:, 0], y=principal_components[:, 1], hue=df['Cluster'], palette='viridis', alpha=0.5)
plt.title('Customer Clusters (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```
